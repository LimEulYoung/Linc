```python
s = '  first star   :   second star   :    third star  '
s = s.split(':')
a = []
for n in s:
    if n != '' :
        a.append( " ".join(n.split()))
print a
```

#[문제 5]
####다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

###[문제 5 - 1]
addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addall(x):
    a = 0
    for n in x:
        a = a + int(n)
    return a

print addall([1])
print addall([1,2,3,4,5,6,7,8,9])
```

```python
def addall(x):
    a = [n for n in x]
    return sum(a,0)

print addall([1])
print addall([1,2,3,4,5,6,7,8,9])
```

###[문제 5 - 2]
addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addallodd(x):
    a = 0
    for n in x:
        if int(n)%2 != 0:
            a = a + int(n)
    return a

print addallodd([1])
print addallodd([1,2,3,4,5,6,7,8,9])
```

```python
def addallodd(x):
    a = [n for n in x if int(n)%2 != 0]
    return sum(a,0)

print addallodd([1])
print addallodd([1,2,3,4,5,6,7,8,9])
```

#implement 2 : 네이버

```python
import string
p = string.punctuation
    

import urllib2
source = urllib2.urlopen("http://www.naver.com").read()

for i in range(0, source.count('<!--')):
    if source.find("<!--") != -1:
        source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

for i in range(0,source.count("<")):
    if source.find("<") != -1:
        source = source.replace(source[source.find("<"):source.find(">")+1:],"")

words = source.split()
words = ' '.join(words)

for j in words:
    for i in p:
        if words.find(i) > 0:
            words =words.replace(words[words.find(i)],'')
            
words = words.split()

b ={}
b.clear()
for i in words:

    if(b.has_key(i)):
        b.update({i:(b.get(i)+1)})
    else:
        b.update({i:1})

for i in b:
    print '(',i,':',b.get(i),')'
```
